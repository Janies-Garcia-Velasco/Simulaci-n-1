```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def aprox(m):
  lista=[]
  for i in range (m):
    u=rnd.random()
    x=((u**0.135)-(1-u)**0.135)/(0.1975)
    lista.append(x)
  return lista
```

```python
G=aprox(5000)
```

```python
media=np.mean(G)
```

```python
varianza=np.var(G)
```
