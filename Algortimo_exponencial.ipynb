```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
lam = 2  # Usamos lam como lambda

def expo(lam, n):
  L=-1/lam
  lista=[]
  for i in range (n):
    u=rnd.random()
    x=L*np.log(u)
    lista.append(x)
  return lista
```

```python
G=expo(0.5,10000)
```

```python
media=np.mean(G)
```

```python
varianza=np.var(G)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y= cdf(G)
```

```python
plt.step(X,Y)
```
