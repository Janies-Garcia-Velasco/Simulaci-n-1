```python
import numpy as np
import random
import matplotlib.pyplot as plt

nums=[]
a=0
c=1
b=2/(c-a)

for i in range(1000):
  x= random.random() #siendo x el numero aleatorio que siga una distribucion triangular

  if x<=a:
    ale=0
    nums.append(ale)
  elif a<x<=c:
    ale=np.sqrt(x*(b-a)*(c-a))+a
    nums.append(ale)
  elif c<x<=b:
    ale=b-np.sqrt()
    nums.append(ale)
  elif x>b:
    ale=1
    nums.append(ale)

plt.hist(nums,bins=100)
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

nums=[]
a=0
c=3/4
b=1

for i in range(1000):
  x= random.random() #siendo x el numero aleatorio que siga una distribucion triangular

  if x<a:
    ale=0
    nums.append(ale)
  elif a<x<=c:
    ale=(5/4)+np.sqrt((25/16)-2*x)
    nums.append(ale)
  elif c<x<=b:
    ale=1+4*x
    nums.append(ale)
  elif x>b:
    ale=1
    nums.append(ale)

plt.hist(nums,bins=100)
plt.show()
```
