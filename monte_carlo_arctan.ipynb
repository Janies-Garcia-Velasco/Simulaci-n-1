```python
#intervalo
a=0
b=1
N=50 # numero de dardos

#Generamos  N puntos aleatorios uniformemente distribuidos en [a,b]
x=np.random.uniform(a,b,N)
y=np.random.uniform(0,1,N)
#evaluando los puntos en la funcion
f_x=np.sqrt(np.arctan(x))

punt_aceptacion=y<f_x

#calculando la integral
int_estimada=(b-a)*np.sum(punt_aceptacion/N)
#imprimimos el resultado
print(f"La estimacion de la integral es: {int_estimada}")
```

```python
#graficacion
plt.figure(figsize=(10,6))
plt.scatter(x[punt_aceptacion],y[punt_aceptacion], color="blue", alpha=0.5, label="puntos aceptados")
plt.scatter(x[~punt_aceptacion],y[~punt_aceptacion], color="green", alpha=0.5, label="puntos rechazados")

#grafica de la funcion para una comparacion
x_vals=np.linspace(a,b,500)
y_vals=np.sqrt(np.arctan(x_vals))
plt.plot(x_vals, y_vals, color="red", label="sqrt(arctan(x))")

plt.xlabel("x")
plt.ylabel("y")
plt.title("estimacion de la integral usando el metodo de Monte Carlo")
plt.legend()
plt.show()
```
