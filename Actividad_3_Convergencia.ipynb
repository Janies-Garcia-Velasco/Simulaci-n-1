```python
import tabulate

xn=6#int(input("escriba un numero semilla: "))
a=5#int(input("escriba un valor para a "))
c=5#int(input("escriba un valor para c: "))
m=16#int(input("escriba un valor para m: "))
n=20


#lista
resultados=[]
#para la tabla
for i in range(n):

  funcion=(a*xn)+c

  funmod=funcion%m

  num_ale=xn/m


  #almacenar resultados en una lista
  resultados.append([i+1,xn,funcion,funmod,num_ale])

  xn=funmod

#para imprimir tabla
tabla=["N","Xn","resultado de funcion","modulo respecto m","cociente respecto m"]
print(tabulate.tabulate(resultados, headers=tabla, tablefmt="grid"))
```
