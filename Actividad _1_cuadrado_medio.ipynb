# Cuadrado medio


```python
import sys

x=int(input("escriba un numero entero de 4 digitos: "))
n=1

while n<10:
  num=str(x)
  if len(num)==4:
    x1=x**2
    num1=str(x1)
    if len(num1)==8:
      numero=list(num1)
      xlist=numero[2:6]
      numfin=int("".join(xlist))
      x=numfin
      print(x)
      n=n+1
    else:
      numero=list(num1)
      xlist=numero[1:5]
      numfin=int("".join(xlist))
      x=numfin
      print(x)
      n=n+1
  else:
    print("el numero ingresado no tiene los digitos solicitados")
    n=10

```
