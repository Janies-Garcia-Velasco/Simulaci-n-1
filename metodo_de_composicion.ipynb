```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parámetros para las dos distribuciones normales
mu1, sigma1, w1 = 0, 1, 0.5  # Media, desviación estándar, peso para la primera distribución normal
mu2, sigma2, w2 = 3, 1, 0.5  # Media, desviación estándar, peso para la segunda distribución normal

# Función de densidad de probabilidad (PDF) de la mezcla
def pdf(x):
    return w1 * norm.pdf(x, mu1, sigma1) + w2 * norm.pdf(x, mu2, sigma2)

# Función de distribución acumulativa (CDF) de la mezcla
def cdf(x):
    return w1 * norm.cdf(x, mu1, sigma1) + w2 * norm.cdf(x, mu2, sigma2)
```

```python
# Graficamos la PDF
x = np.linspace(-5, 8, 1000)
y = pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='PDF de la mezcla', color='blue')
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficamos la CDF
y_cdf = cdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_cdf, label='CDF de la mezcla', color='green')
plt.title('Función de Distribución Acumulativa (CDF)')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generamos muestras utilizando el método de composición
n_muestras = 10000
muestras = np.zeros(n_muestras)
u = np.random.rand(n_muestras)  # Números aleatorios distribuidos uniformemente
```

```python
# Utilizamos el método de composición para generar las muestras
for i in range(n_muestras):
    if u[i] < w1:
        muestras[i] = np.random.normal(mu1, sigma1)
    else:
        muestras[i] = np.random.normal(mu2, sigma2)
```

```python
# Graficamos el histograma de las muestras generadas
plt.figure(figsize=(10, 6))
plt.hist(muestras, bins=50, density=True, alpha=0.6, color='gray', edgecolor='black')
plt.plot(x, y, label='PDF de la mezcla', color='blue', linewidth=2)
plt.title('Histograma de las Muestras')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
